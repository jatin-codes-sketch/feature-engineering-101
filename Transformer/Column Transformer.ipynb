```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\Lenovo\OneDrive\Documents\covid.csv')
df.sample(5)

df1=df.drop('City',axis=1)
df1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df1.iloc[:,1:],df1.iloc[:,-1:],test_size=0.2,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
transformer=ColumnTransformer(transformers=[
    ('trans1',OneHotEncoder(sparse_output=False,drop='first'),['Gender']),
    ('trans2',OneHotEncoder(sparse_output=False,drop='first'),['Fever']),
    ('trans3',OrdinalEncoder(categories=[['Low','Mild','High']]),['Cough']),
    ('trans4',OneHotEncoder(sparse_output=False,drop='first'),['State'])
],remainder='passthrough')
```

```python
X_trained=transformer.fit_transform(X_train)
```

```python
# Assuming `X_trained` is the result of your transformation (e.g., from ColumnTransformer or OneHotEncoder)
# Get the feature names after transformation
feature_names = transformer.get_feature_names_out(input_features=X_train.columns)

# Create the final DataFrame with the correct column names
final = pd.DataFrame(X_trained, columns=feature_names)

# View the final DataFrame
final
```

```python
#  for writing in more modern and readable approach
from sklearn.compose import make_column_transformer
```

```python
preprocessor=make_column_transformer(
    (OneHotEncoder(sparse_output=False,drop='first'),['Gender']),
    (OneHotEncoder(sparse_output=False,drop='first'),['Fever']),
    (OrdinalEncoder(categories=[['Low','Mild','High']]),['Cough']),
    (OneHotEncoder(sparse_output=False,drop='first'),['State']),
    remainder='passthrough'
)
```

```python
X_train_preprocessed=preprocessor.fit_transform(X_train)
X_test_preprocessed=preprocessor.transform(X_test)
```

```python
feature_names=preprocessor.get_feature_names_out(input_features=X_train.columns)

fin=pd.DataFrame(X_train_preprocessed,columns=feature_names)
fin
```
