```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df=df.drop(columns=['species'])
df.head()
```

```python
df.shape
```

```python
sns.kdeplot(df['sepal_length'])
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
X_train,y_train,X_test,y_test=train_test_split(df,df,random_state=42,test_size=0.2)
```

```python
mm=MinMaxScaler()
mm.fit_transform(X_train)

X_train_scaled=mm.transform(X_train)
X_test_scaled=mm.transform(X_test)
```

```python
X_train_scaled
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_train_scaled.head()
```

```python
print(X_train.describe())
print(X_train_scaled.describe())
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,5))


# before scaling
ax1.scatter(X_train['sepal_length'],X_train['sepal_width'])
ax1.set_title('Before Scaling')

ax1.set_xlabel('sepal_length')
ax1.set_ylabel('sepal_width')


# after scaling
ax2.scatter(X_train_scaled['sepal_length'],X_train_scaled['sepal_width'],color='red')
ax2.set_title('After Scaling')
ax2.set_xlabel('sepal_length')
ax2.set_ylabel('sepal_width')
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,5))

# before scaling
sns.kdeplot(X_train['petal_length'],ax=ax1)
sns.kdeplot(X_train['petal_width'],ax=ax1)
ax1.set_title('Before Scaling')
ax1.set_xlabel('petal_length')
ax1.set_ylabel('petal_width')

# after scaling
sns.kdeplot(X_train_scaled['petal_length'],ax=ax2)
sns.kdeplot(X_train_scaled['petal_width'],ax=ax2,color='red')
ax2.set_title('After Scaling')
ax2.set_xlabel('petal_length')
ax2.set_ylabel('petal_width')
plt.show()
```
