```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv(r'C:\Users\Lenovo\Downloads\hotel_bookings.csv')
df.head(5)
```

```python
df.info()
```

```python
df['Booking_DateTime']=pd.to_datetime(df['Booking_DateTime'])
df['date']=df['Booking_DateTime'].dt.date
df['time']=df['Booking_DateTime'].dt.time
df
```

#### 1) extract year 

```python
df['date']=pd.to_datetime(df['date'])
df['date_year']=df['date'].dt.year
df
```

#### 2) extract months with months names

```python
df['month']=df['date'].dt.month
df['month_name']=df['date'].dt.month_name()
df
```

#### 3) extract days with date names and day of week

```python
df['date_day']=df['date'].dt.day
df['date_day_name']=df['date'].dt.day_name()
df['dayweek']=df['date'].dt.dayofweek
df
```

```python
# is weekend ?

df['date_is_weekend']=np.where(df['date_day_name'].isin(['Sunday','Saturday']),1,0)

df
```

#### extract week of the year 

```python
df['date_week']=df['date'].dt.isocalendar().week
df
```

#### extract quarters 

```python
df['quarter']=df['date'].dt.quarter
df
```

#### extract semesters 

```python
df['semester']=np.where(df['quarter'].isin([1,2]),1,2)
df
```
