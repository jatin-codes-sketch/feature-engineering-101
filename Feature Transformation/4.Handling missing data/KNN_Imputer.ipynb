```python
import pandas as pd 
import numpy as np 

from sklearn.model_selection import train_test_split 
from sklearn.impute import SimpleImputer,KNNImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df=df[['age','fare','pclass','survived']]
df
```

```python
df.isnull().mean()*100
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.2,random_state=32)
```

```python
kkn=KNNImputer(n_neighbors=3,weights="uniform")  # n_neighbors: int default=5 and weight=['uniform','distance'] default='uniform' 
```

```python
X_train_trf=kkn.fit_transform(X_train)
X_test_trf=kkn.transform(X_test)
```

```python
pd.DataFrame(X_test_trf,columns=X_train.columns)
```

```python
lr=LogisticRegression()
lr.fit(X_train_trf,y_train)
y_pred=lr.predict(X_test_trf)
accuracy_score(y_test,y_pred)
```

```python
si=SimpleImputer()
X_train_si=si.fit_transform(X_train)
X_test_si=si.transform(X_test)
lr.fit(X_train_si,y_train)
y_pred=lr.predict(X_test_si)
accuracy_score(y_test,y_pred)
```
